# Lets fill the missing data


```python
df['name'] = df['name'].fillna('unknown')
```

```python
df['host name'] = df['host name'].fillna('Unknown')
```

```python
df.fillna({'host_identity_verified':'unconfirmed'}, inplace=True)
```

```python
df.fillna({'number of reviews':0}, inplace=True)
```

```python
df.fillna({'country':'unknown', 'country code':'unknown', 'instant_bookable':'False', 'neighbourhood':'unknown','neighbourhood group':'unknown'},inplace=True)
```

```python
df.dropna(subset=['lat', 'long'], inplace=True)
```

```python
df['cancellation_policy'] = df['cancellation_policy'].fillna('Not specified')
```

```python
# num_cols = [
#     'minimum nights',
#     'number of reviews',
# ]

# for col in num_cols:
#     df[col].fillna(df[col].median(), inplace=True)
```

```python
df['minimum nights'] = df['minimum nights'].median()
```

```python
df['price'] = (
    df['price']
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)


df['price'] = df['price'].median()
```

```python

df['service fee'] = (
    df['service fee']
    .str.replace("$", '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)


df['service fee'] = df['service fee'].median()
```

```python
df['Construction year'] = df['Construction year'].median()
```

```python
df.isna().sum()
```

# Lets clean individual column

```python
df['country code'] = df['country code'].str.upper()
```

```python
df.head()
```

```python
df['instant_bookable'].apply(lambda a: 1 if a == True else 0)
```

```python
df['instant_bookable'] = df['instant_bookable'].astype(bool)
df['number of reviews'] = df['number of reviews'].astype(int)
df['minimum nights'] = df['minimum nights'].astype(int)
```

```python
# df.reset_index(drop=True, inplace=True)

df = df.reset_index(names='id')

```
